# Исследование рынка копьютерных игр #

Интернет-магазина "Стримчик" продает по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр за 1980-2016 гг., в которых содержится информация о платформах, играх, жанрах компьютерных игр, годах релиза, оценках пользователей и критиков, рейтинги от ассоциации ERSB, а также продажи в различных регионах. 

На основании этих данных необходимо спрогнозировать кампанию на следующий год. 

Целью данного исследования является выявление закономерностей, определеющих успешность игры, что позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

## Предобработка данных

### Преобразование названий столбцов

```python
# приведем названия столбцов к нижнему регистру
data.columns = data.columns.str.lower()
data.columns
```

### Преобразование данных в нужные типы

В столбце с годом релиза игр приведем данные к целочисленному типу, сведения о продажах округлим до десятых, в  столбце 'user_score' встречается аббревиатура `tbd`, которая означает, что оценка пользователей еще не определена, поэтому данное значение можно считать пропуском, заменим его на 'NaN', значения типа 'object' приведем к типу 'float64', а в столбце 'Year_of_Release' приведем данные к целочисленному типу.

```python
# преобразование данных
data['year_of_release'] = data['year_of_release'].astype('Int64')
data['na_sales'] = data['na_sales'].round(1)
data['eu_sales'] = data['eu_sales'].round(1)
data['jp_sales'] = data['jp_sales'].round(1)
data['other_sales'] = data['other_sales'].round(1)
data['user_score'] = data['user_score'].replace('tbd', np.NaN)
data['user_score'] = data['user_score'].astype('float64')
# проверка типов данных
data.dtypes
```

### Обработка пропусков

```python
# подсчет общего числа пропусков
data.isna().sum()
```

```python
# подсчет доли пропусков
data.isna().mean()
```

Найдены пропущенные значения в столбах с названием игр, годом релиза, жанром, оценками пользователей и критиков, а также рейтингом. В пропусках в столбцах с оценкой критиков и пользователей поставим заглушки '9999', которые в дальнейшем исследовании будем исключать из анализа. Удалим строки с пропущенными значениями в столбцах 'name' и 'year_of_release', поскольку они занимают очень маленькую долю от общего числа данных и не повлияют на результат исследования, а пропуски в столбцах с рейтингом заполним значением 'unknown', поскольку он мог быть просто не определен по каким-либо неизвестным нам причинам.

```python
data['critic_score'] = data['critic_score'].fillna(9999)
data['user_score'] = data['user_score'].fillna(9999)
data = data.dropna(subset=['name', 'year_of_release'])
data['rating'] = data['rating'].fillna(value='unknown')
```

```python
# проверим число пропусков
data.isna().sum()
```

ESRB - это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию.

```python
# для удобства добавим в таблицу столбец с категоризацией рейтинга
def rating_function(rating):
    try:
        if rating == 'EC':
            return 'для детей младшего возраста'
        elif rating == 'E' or rating == 'K-A':
            return 'для всех'
        elif rating == 'E10+':
            return 'для всех старше 10 лет'
        elif rating == 'T':
            return 'подросткам 13—19 лет'
        elif rating == 'M':
            return 'от 17 лет'
        elif rating == 'AO':
            return 'только для взрослых'
        elif rating == 'RP':
            return 'рейтинг ожидается'
        elif rating == 'unknown':
            return 'неизвестен'
    except:
        return 'проверить данные'
        
data['rating_category'] = data['rating'].apply(rating_function)
```

### Изучение изменения продаж по платформам

```python
# построим график распределения продаж по платформам
sales_platform = data.pivot_table(index='platform', values='all_sales', aggfunc='sum').sort_values(by='all_sales', ascending=False)
sales_platform.plot(kind='bar', grid=True, figsize=(7,5))
plt.xlabel("платформы")
plt.ylabel("суммарные продажи")
plt.title("Распределение продаж по платформам")
plt.show()
```

Лидирующими по продажам являются платформы PS2, X360, PS3, Wii, DS, PS.

```python
# выберем платформы с максимальным количеством продаж
sales_platform = sales_platform.head(6)
# создадим из них список
top_sales_platform = sales_platform.head(6).index.to_list()
# сформируем сводную таблицу
top_sales_platform_pivot = data.query('platform in @top_sales_platform').pivot_table(index='year_of_release', 
                                                                         columns='platform', 
                                                                         values='all_sales', 
                                                                         aggfunc='sum')
# построим график распределения продаж на платформах по годам 
top_sales_platform_pivot.plot(grid=True, figsize=(8,5))
plt.xlabel("год релиза")
plt.ylabel("суммарные продажи")
plt.title("Распределение продаж по платформам")
plt.show()
```

По графику мы видим, что жизненный цикл платформ составляет около 10 лет.

### Определение актуального периода

Рассмотрим продажи на платформах за последние 10 лет

```python
# построим график распределения продаж на платформах c 2005 года
data.query('year_of_release >= 2005').groupby(['platform', 'year_of_release'])['all_sales'].sum().unstack('platform').plot(grid=True, figsize=(12,12))
plt.xlabel("год релиза")
plt.ylabel("суммарные продажи")
plt.title("Распределение продаж по платформам")
plt.show()
```

На графике наблюдается спад актульности многих платформ, по некоторым продажи падают, по другим исчезают совсем. Одновременно с этим появляются новые платформы. Потенциально прибыльными платформами будет считать платформы, которые появились относительно недавно и находятся в первой половине своего жизненного цикла.

```python
# получение среза данных за актуальный период
actual_data = data.query('year_of_release >= 2014')
# выберем несколько потенциально прибыльных платформ
profit_platform = actual_data.pivot_table(index='platform', values='all_sales', aggfunc='sum').sort_values(by='all_sales', ascending=False)
top_profit_platform = profit_platform.head(5)
top_profit_platform
```

Мы определели актуальный для нас период с платформами которые появились не так давно и имеют хорошие продажи.

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```python
# построим гистограммы распределения пользовательских рейтингов платформ Xbox One и PC
user_score_XOne = actual_data.query('platform == "XOne" and user_score != 9999')['user_score']
user_score_PC = actual_data.query('platform == "PC" and user_score != 9999')['user_score']
plt.figure(figsize=(9,7))
ax=user_score_XOne.plot(kind='hist', y='user_score', bins=30, alpha=0.5, label='XOne')
user_score_PC.plot(kind='hist', y='user_score', bins=30, alpha=0.5, label='PC', ax=ax, grid=True, legend=True)
plt.xlabel("оценки пользователей")
plt.ylabel("частота")
plt.title("Распределение пользовательских рейтингов на платформам Xbox One и PC ")
plt.show()
```

```python
# посчитаем средние пользовательские рейтинги платформ Xbox One и PC
user_score_XOne = actual_data.query('platform == "XOne" and user_score != 9999')['user_score'].mean()
user_score_PC = actual_data.query('platform == "PC" and user_score != 9999')['user_score'].mean()
display(user_score_XOne)
display(user_score_PC)
```

Но - средние пользовательские рейтинги платформ Xbox One и PC равны  
Н1 - средние пользовательские рейтинги платформ Xbox One и PC не равны


```python
# проверим гипотезу
user_score_XOne = actual_data.query('platform == "XOne" and user_score != 9999')['user_score']
user_score_PC = actual_data.query('platform == "PC" and user_score != 9999')['user_score']

alpha = 0.05
result = st.ttest_ind(user_score_XOne, user_score_PC, equal_var=False)
print('р-значение:', result.pvalue)

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные

```python
# построим гистограммы распределения пользовательских рейтингов жанров Action и Sports
user_score_action = actual_data.query('genre == "Action" and user_score != 9999')['user_score']
user_score_sports = actual_data.query('genre == "Sports" and user_score != 9999')['user_score']
plt.figure(figsize=(9,7))
ax=user_score_action.plot(kind='hist', y='user_score', bins=30, alpha=0.5, label='Action')
user_score_sports.plot(kind='hist', y='user_score', bins=30, alpha=0.5, label='Sports', ax=ax, grid=True, legend=True)
plt.xlabel("оценки пользователей")
plt.ylabel("частота")
plt.title("Распределение пользовательских рейтингов жанров Action и Sports")
plt.show()
```

```python
# посчитаем средние пользовательские рейтинги жанров Action и Sports
user_score_action = actual_data.query('genre == "Action" and user_score != 9999')['user_score'].mean()
user_score_sports = actual_data.query('genre == "Sports" and user_score != 9999')['user_score'].mean()
display(user_score_action)
user_score_sports
```

Но - средние пользовательские рейтинги жанров Action и Sports равны  
Н1 - средние пользовательские рейтинги жанров Action и Sports не равны

```python
# проверим гипотезу

user_score_action = actual_data.query('genre == "Action" and user_score != 9999')['user_score']
user_score_sports = actual_data.query('genre == "Sports" and user_score != 9999')['user_score'] 
alpha = 0.05
result = st.ttest_ind(user_score_action, user_score_sports, equal_var=False)
print('р-значение:', result.pvalue)

if result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

**Вывод**

Итак, мы проверили гипотезы о равенстве средних пользовательских рейтингов платформ Xbox One и PC и о равенстве средних пользовательских рейтингов жанров Action и Sports. Для проверки был применен метод scipy.stats.ttest_ind поскольку наши выборки независимые. Для первой гипотезы у нас не было оснований, чтобы ее отклонить, поскольку р-уровень значимости выше порогового значения. Обратная картина сложилась для второй гипотезы и нам пришлось ее отвергнуть, поскольку р-уровень значимости был намного ниже критического.
